```python
from matplotlib import pyplot as plt
import numpy as np

def mul_entropy(p, **kwargs):
    p = np.array(p)
    return np.sum(-p * np.log2(p), **kwargs)
def entropy(p):
    return -p * np.log2(p) - (1 - p) * np.log2(1 - p)
p_values = np.linspace(0.01, 0.99, 100)
entropy_values = entropy(p_values)
plt.plot(p_values, entropy_values)
plt.title('Entropy of a Bernoulli Distribution')
plt.xlabel('Probability of Success (p)')
plt.ylabel('Entropy (H)')
plt.grid()
plt.show()

# 3d plot with two probabilities
from plotly import graph_objects as go
p1_values = np.linspace(0.01, 0.99, 50)
p2_values = np.linspace(0.01, 0.99, 50)
p1_grid, p2_grid = np.meshgrid(p1_values, p2_values)
p3_values = 1 - p1_grid - p2_grid
entropy_grid = mul_entropy([p1_grid, p2_grid, p3_values], axis = 0)
fig = go.Figure(data=[go.Surface(z=entropy_grid, x=p1_grid, y=p2_grid)])
fig.update_layout(title='Joint Entropy of Two Bernoulli Distributions',
                  scene=dict(xaxis_title='Probability of Success (p1)',
                             yaxis_title='Probability of Success (p2)',
                             zaxis_title='Joint Entropy (H)'))
fig.show()
```
